# Marketing KPI Analysis

## Objective
This notebook calculates the key marketing performance indicators (KPIs) required for business analysis and dashboard development.

### KPIs Covered
- Total Revenue
- Total Cost
- Total Clicks
- Total Leads
- Total Conversions
- Average ROI
- ROAS
- CPC
- CPL

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/processed/marketing_campaign_cleaned.csv")
```

```python
df.head()
```

```python
print("Rows and Columns:", df.shape)
print("\nColumn Names:")
print(df.columns.tolist())
```

```python
total_revenue = df["Revenue_USD"].sum()

print("Total Revenue: $", round(total_revenue, 2))
```

```python
total_cost = df["Cost_USD"].sum()

print("Total Cost: $", round(total_cost, 2))
```

```python
total_clicks = df["Clicks"].sum()

print("Total Clicks:", total_clicks)
```

```python
total_leads = df["Leads"].sum()

print("Total Leads:", total_leads)
```

```python
total_conversions = df["Conversions"].sum()

print("Total Conversions:", total_conversions)
```

```python
roas = total_revenue / total_cost

print("ROAS:", round(roas, 2))
```

```python
cpc = total_cost / total_clicks

print("Cost Per Click (CPC): $", round(cpc, 2))
```

```python
cpl = total_cost / total_leads

print("Cost Per Lead (CPL): $", round(cpl, 2))
```

```python
conversion_rate = (total_conversions / total_clicks) * 100

print("Conversion Rate:", round(conversion_rate, 2), "%")
```

```python
average_roi = df["ROI"].mean()

print("Average ROI:", round(average_roi, 2))
```

```python
kpi_summary = pd.DataFrame({
    "KPI": [
        "Total Revenue",
        "Total Cost",
        "Total Clicks",
        "Total Leads",
        "Total Conversions",
        "ROAS",
        "CPC",
        "CPL",
        "Conversion Rate (%)",
        "Average ROI"
    ],
    "Value": [
        round(total_revenue, 2),
        round(total_cost, 2),
        total_clicks,
        total_leads,
        total_conversions,
        round(roas, 2),
        round(cpc, 2),
        round(cpl, 2),
        round(conversion_rate, 2),
        round(average_roi, 2)
    ]
})

kpi_summary
```

```python
kpi_summary.to_csv("../reports/kpi_summary.csv", index=False)

print("KPI Summary saved successfully!")
```
